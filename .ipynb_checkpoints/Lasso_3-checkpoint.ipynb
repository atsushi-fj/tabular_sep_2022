# 正則化項による回帰

## ライブラリのインポート

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import Lasso
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
```

## データセットのインポート

```python
train_df = pd.read_csv('train.csv', dtype={1:str}, parse_dates=[1])
test_df = pd.read_csv('test.csv', dtype={1:str}, parse_dates=[1])
```

### 年、月、日のデータの抽出

```python
train_df['date'].nunique()
```

```python
train_df['year'] = train_df['date'].dt.year
test_df['year'] = test_df['date'].dt.year
train_df['month'] = train_df['date'].dt.month
test_df['month'] = test_df['date'].dt.month
train_df['day'] = train_df['date'].dt.day
test_df['day'] = test_df['date'].dt.day
```

```python
train_df = train_df.drop('date', axis=1)
test_df = test_df.drop('date', axis=1)
```

```python
train_df.head()
```

```python
import seaborn as sns
sns.pairplot(train_df)
```

### 特徴量と目的変数

```python
y_col = 'num_sold'
X = train_df.loc[:, train_df.columns!=y_col]
y = train_df[y_col]
```

```python
X = X.drop('row_id', axis=1)
index_test = test_df['row_id']
test_df = test_df.drop('row_id', axis=1)
# 数値カラムの取得
numeric_cols = X.select_dtypes(include=np.number).columns.to_list()
numeric_cols
```

### one-hot-encoding

```python
X['country'].unique()
```

```python
X['store'].unique()
```

```python
X['product'].unique()
```

```python
X = pd.get_dummies(X, drop_first=True)
test_df = pd.get_dummies(test_df, drop_first=True)
```

### 標準化

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
scaler = StandardScaler()
X_train[numeric_cols] = scaler.fit_transform(X_train[numeric_cols])
X_test[numeric_cols] = scaler.transform(X_test[numeric_cols]
```

## 最適なLasso

```python
alphas = np.logspace(-3, 3)# 1e-3から1e3までを50個に分割したリスト
smape_list = []
coefs = []
for alpha in alphas:
    model = Lasso(alpha)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    smape = mean_absolute_percentage_error(y_test, y_pred)
    smape_list.append(smape)
    coefs.append(model.coef_)
```

```python
plt.plot(alphas, smape_list)
plt.xscale('log')
plt.xlabel('lambda')
plt.ylabel('test SMAPE')
```

```python
plt.plot(alphas, coefs)
plt.xscale('log')
plt.xlabel('lambda')
plt.ylabel('standardized coefficients')
```

## モデルの学習

```python
lasso = Lasso(alphas[np.argmin(smape_list)])
lasso.fit(X_train, y_train)
```

```python
y_pred = lasso.predict(X_test)
smape = mean_absolute_percentage_error(y_test, y_pred)
smape
```

## テストデータに対する予測

```python
test_df[numeric_cols] = scaler.transform(test_df[numeric_cols])
```

```python
X_test = test_df.values
```

```python
y_pred = lasso.predict(X_test)
```

```python
y_pred
```
